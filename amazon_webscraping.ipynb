```python
from autoscraper import AutoScraper
```

```python
amazon_url="https://www.amazon.in/s?k=oneplus10R"
wanted_list=["₹42,999","OnePlus 10R 5G (Forest Green, 12GB RAM, 256GB Storage, 80W SuperVOOC) ", "  218 "]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_a4dt':'Title','rule_6xmz':'Price'})
scraper.keep_rules(['rule_a4dt','rule_6xmz'])
scraper.save("Amazon_Search")
```

```python
results=scraper.get_result_similar("https://www.amazon.in/s?k=mi+note10",group_by_alias=True)
results['Title']
```
